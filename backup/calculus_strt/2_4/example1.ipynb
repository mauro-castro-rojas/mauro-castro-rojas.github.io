```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

x = sp.Symbol('x')
function = x**3 - 5*x + 6
x_vals = np.linspace(0.5, 1.5, 200)

lambda_function = sp.lambdify(x, function, 'numpy')
y_vals = lambda_function(x_vals)


fig, ax = plt.subplots(figsize=(4, 4))
ax.plot(x_vals, y_vals, label=r"$f(x)=x^3 - 5*x + 6$")
ax.plot(x_vals, np.full_like(x_vals, 2.2), label="h_line at y = 2.2", linestyle='--', color='red')
ax.plot(x_vals, np.full_like(x_vals, 1.8), label="h_line at y = 1.8", linestyle='--', color='red')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()


```
