## <span style="color:DodgerBlue">Derivatives of polynomials and Exponential Functions</span>

### <span style="color:Tomato">Constant functions</span>

$$ \frac{d}{dx} (c) = 0$$



```python
x = sp.Symbol('x')
sample_data = np.linspace(-2, 2, 100) 
constant_values = np.full_like(sample_data, 2)

f_prime = sp.diff(sample_data[0], x)
print("Derivative at any point:", f_prime)

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(sample_data, constant_values, label='linear') 
```

### <span style="color:Tomato">Power rule</span>

If $n$ is any real number, then

$$ \frac{d}{dx} (x^n) = n x^{n-1}$$

```python
function = x**(2/3)
y_vals = safe_lambda_function(sample_data, function)

plt.plot(sample_data, y_vals)
```
